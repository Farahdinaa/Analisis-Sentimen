```python
from google_play_scraper import reviews, Sort
import pandas as pd
```

```python
package_name = "com.disney.disneyplus"
all_reviews = []
```

```python
for score in [1, 2, 3, 4, 5]:
    total_reviews = 0
    while total_reviews < 2000:
        result, _ = reviews(
            package_name,
            lang='en',
            country='us',
            sort=Sort.NEWEST,
            count=100,
            filter_score_with=score
        )
        all_reviews.extend(result)
        total_reviews += len(result)
```

```python
reviews_df = pd.DataFrame(all_reviews)
print(f"Jumlah ulasan: {len(reviews_df)}")
```

```python
reviews_df.to_csv("disney_reviews.csv", index=False)
```
